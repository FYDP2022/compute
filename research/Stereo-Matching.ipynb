# Stereo Matching for Lawny

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import math

lawn = mpimg.imread('assets/lawn.png')

plt.imshow(lawn)
```

```python
def color_mask(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    mask = cv2.inRange(hsv, np.array([65, 0.4, 0.3]), np.array([160, 1.0, 1.0]))
    return cv2.bitwise_and(image, image, mask=mask)

def pipeline(source):
    masked = color_mask(source)
    hsv = cv2.cvtColor(masked, cv2.COLOR_RGB2HSV)
    for i in range(hsv.shape[0]):
        for j in range(hsv.shape[1]):
            if hsv[i][j][0] >= 65 and hsv[i][j][0] <= 160:
                hsv[i][j][1] = 1.0
                hsv[i][j][2] = 1.0
    normalized = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    gray = cv2.cvtColor(normalized, cv2.COLOR_RGB2GRAY)
    plt.figure()
    plt.imshow(gray, cmap='gray')
#     grad_x = cv2.Sobel(gray, cv2.CV_16S, 1, 0, ksize=5, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
#     plt.figure()
#     plt.imshow(grad_x)
#     grad_y = cv2.Sobel(gray, cv2.CV_16S, 0, 1, ksize=5, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
#     plt.figure()
#     plt.imshow(grad_y)
    filtered = None
    for i in range(13):
        kernel = cv2.getGaborKernel((8, 8), 2, i * 15 * math.pi / 180.0, 5, 0, 0, ktype=cv2.CV_32F)
        plt.figure()
        plt.imshow(kernel)
        tmp = cv2.filter2D(gray, cv2.CV_8UC3, kernel)
        if filtered is None:
            filtered = tmp
        else:
            filtered = cv2.addWeighted(filtered, i / (i + 1), tmp, 1 / (i + 1), 0.0)
        plt.figure()
        plt.imshow(tmp)
    plt.figure()
    plt.imshow(filtered)
#     blur = cv2.GaussianBlur(filtered, (7, 7), 0)
#     plt.figure()
#     plt.imshow(blur, cmap='gray')
    mask = cv2.threshold(filtered, 0, 3, cv2.THRESH_BINARY)[1]
    plt.figure()
    plt.imshow(mask, cmap='gray')
    output = cv2.bitwise_and(source, source, mask=mask)
    plt.figure()
    plt.imshow(output)

pipeline(lawn)
```
